```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Histogram with 5 bins
plt.figure(figsize=(8,5))
plt.hist(df['total_bill'], bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Total Bill (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill (5 Bins)")
plt.show()

# Histogram with 20 bins
plt.figure(figsize=(8,5))
plt.hist(df['total_bill'], bins=20, color='orange', edgecolor='black')
plt.xlabel("Total Bill (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill (20 Bins)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', data=df)
plt.xlabel("Day")
plt.ylabel("Total Bill (USD)")
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
# Filter Sunday data
sun_data = df[df['day'] == 'Sun']['total_bill']

# Calculate Q1, Q3, and IQR
Q1 = sun_data.quantile(0.25)
Q3 = sun_data.quantile(0.75)
IQR = Q3 - Q1

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q3, IQR, lower_bound, upper_bound
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="time",
    hover_data=["day", "size"],
    title="Total Bill vs Tip (Colored by Time)"
)

fig.show()
```

There is a positive relationship between total_bill and tip:
As the total bill increases, the tip generally increases.
This suggests people tend to tip proportionally.
Lunch vs Dinner differences:
Dinner points are more spread out and include higher bill values.
Lunch transactions are more tightly clustered at lower bill amounts.

This indicates that dinner meals tend to be larger and more variable, leading to higher tips overall.

Interesting data point:
One noticeable point is a very high total_bill (around $50) with a relatively moderate tip.
This is interesting because:
It deviates slightly from the general proportional trend.
It may indicate a lower tip percentage compared to similar bills.
