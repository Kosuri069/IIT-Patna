```python
import pandas as pd

df = pd.read_csv("car_dataset.csv")  # or your file name

df.head()
df.info()
df.describe()
df.shape
```

```python
df["brand"] = df["name"].str.split().str[0]
df["brand"] = df["brand"].str.strip().str.lower()
```

```python
df = df.dropna(subset=["selling_price"])
```

```python
df["mileage"] = df["mileage"].astype(str).str.extract(r'(\d+\.?\d*)')
df["mileage"] = pd.to_numeric(df["mileage"], errors="coerce")
```

```python
df["engine"] = df["engine"].astype(str).str.extract(r'(\d+\.?\d*)')
df["engine"] = pd.to_numeric(df["engine"], errors="coerce")

df["max_power"] = df["max_power"].astype(str).str.extract(r'(\d+\.?\d*)')
df["max_power"] = pd.to_numeric(df["max_power"], errors="coerce")

df["torque"] = df["torque"].astype(str).str.extract(r'(\d+\.?\d*)')
df["torque"] = pd.to_numeric(df["torque"], errors="coerce")
```

```python
numeric_cols = ["mileage", "engine", "max_power", "torque", "seats"]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = ["fuel", "seller_type", "transmission", "owner", "brand"]

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df = df.drop_duplicates()
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df.info()
df.describe()
df.shape
```

```python
from sklearn.metrics import mean_absolute_error

y_true = df["selling_price"]
baseline_prediction = y_true.mean()
y_pred = [baseline_prediction] * len(y_true)

mae = mean_absolute_error(y_true, y_pred)

print("Baseline prediction (mean selling_price):", baseline_prediction)
print("Baseline MAE:", mae)
```

1. Exploration findings
*   The dataset contains 8128 rows and 13 columns.
*   Several columns such as mileage, engine, max_power, and torque were stored as object instead of numeric.
*   Missing values were found in mileage, engine, max_power, torque, and seats.
*   A brand column was derived from the name column and standardized.
*   Duplicate rows were removed.

2. Cleaning actions
*   Dropped rows with null target values in selling_price
*   Extracted numeric values from mileage, engine, max_power, and torque
*   Imputed missing feature values
*   Standardized brand text using strip and lowercase
*   Removed duplicates

3. Baseline model
*   Built a baseline model that predicts the mean selling price for every record
*   Calculated MAE to use as a benchmark before applying machine learning models
