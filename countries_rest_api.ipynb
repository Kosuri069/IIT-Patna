```python
import requests

import requests

# Restrict fields to only what we need
url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries retrieved:", len(data))
    else:
        print("Error:", response.text)

except requests.exceptions.RequestException as e:
    print("Request failed:", e)
```

```python
import pandas as pd

# Extract relevant fields
countries_list = []

for country in data:
    common_name = country.get("name", {}).get("common", "N/A")
    population = country.get("population", 0)
    region = country.get("region", "N/A")
    area = country.get("area", 0)

    countries_list.append({
        "common_name": common_name,
        "population": population,
        "region": region,
        "area": area
    })

# Create DataFrame
df = pd.DataFrame(countries_list)

# Display first 10 rows
df.head(10)
```

```python
# Region with highest total population
region_population = df.groupby("region")["population"].sum()
highest_population_region = region_population.idxmax()

# Country with largest area
largest_country = df.loc[df["area"].idxmax(), "common_name"]

# Print results
print("Region with the highest total population:", highest_population_region)
print("Country with the largest area:", largest_country)
```
