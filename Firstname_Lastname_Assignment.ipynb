```python
import pandas as pd
import requests

# Google Drive file ID
file_id = "1Mlu9r3afSGz5_02Y3wDJoQLpXszQXaIU"

# Direct download URL
url = f"https://drive.google.com/uc?export=download&id={file_id}"

# Download the file
response = requests.get(url)

# Save as CSV
with open("orders.csv", "wb") as f:
    f.write(response.content)

# Load dataset with order_id as index
df = pd.read_csv("orders.csv", index_col="order_id")

# Task 1: Display first and last 5 rows
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

# Task 2: Structural inspection
print("\nDataset Info:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

# Convert order_date to datetime
df["order_date"] = pd.to_datetime(df["order_date"])

# Task 3: Statistical summary
print("\nStatistical Summary:")
print(df.describe())

print("\nMedian unit_price:")
print(df["unit_price"].median())
```
